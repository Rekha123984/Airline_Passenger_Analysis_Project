```python
# Cleaning Unicode Characters & Saving File
import pandas as pd
import unicodedata
from google.colab import files

df=pd.read_csv("Cleaned Airline Dataset.csv")

def normalize_text(text):
  if isinstance(text,str):
    return unicodedata.normalize("NFKD",text).encode("ascii","ignore").decode("utf-8")
  return text

for col in df.select_dtypes(include="object").columns:
  df[col]=df[col].apply(normalize_text).str.strip()

df.to_csv("Airline_Dataset.csv",index=False,encoding="utf-8")

files.download("Airline_Dataset.csv")
```

```python
# Importing the Libraries and Reading the Cleaned Unicode Characters File
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
import numpy as np
import scipy.stats as stats

df=pd.read_csv("Airline_Dataset.csv")
print(df.head())
```

```python
# Filling the Age Column with the Mean Value
x=df["Age"].mean()
df.fillna({"Age":x},inplace=True)
print(df.head())
```

```python
# Dropping the Null Values
df.dropna(inplace=True)
print(df.head())
```

```python
# Correcting the Date Format
df["Departure_Date"]=pd.to_datetime(df["Departure_Date"],format="mixed")
print(df.head())
```

```python
# Dropping the Duplicate
df.drop_duplicates(inplace=True)
print(df.head())
```

```python
# Split ages based on flight status
on_time = df[df["Flight_Status"] == "On Time"]["Age"].dropna()
delayed = df[df["Flight_Status"] == "Delayed"]["Age"].dropna()

# Independent t-test
t_stat, p_val = stats.ttest_ind(on_time, delayed, equal_var=False)

print("T-Statistic:", t_stat)
print("P-Value:", p_val)

# Interpretation + Storytelling
if p_val < 0.05:
    print("Reject H0 → Passenger age differs between On Time and Delayed flights.")
    print("Finding: Passenger age seems to influence delays. Certain age groups might be linked with flight reliability.")
else:
    print("Fail to Reject H0 → Passenger age does not significantly differ between flight statuses.")
    print("Finding: Passenger age has no meaningful relationship with delays. This shows delays are mainly caused by operational issues such as weather, scheduling, or airport congestion, not by passenger demographics.")
```

```python
# Contingency table
contingency = pd.crosstab(df["Airport_Continent"], df["Flight_Status"])

# Chi-square test
chi2, p_val, dof, expected = stats.chi2_contingency(contingency)

print("Chi-Square Statistic:", chi2)
print("P-Value:", p_val)
print("Degrees of Freedom:", dof)

# Interpretation + Storytelling
if p_val < 0.05:
    print("Reject H0 → Flight status depends on the continent of departure.")
    print("Finding: Flight delays/cancellations vary across continents. Some regions face more reliability issues than others, which airlines can investigate.")
else:
    print("Fail to Reject H0 → Flight status is independent of the continent of departure.")
    print("Finding: Flight reliability does not significantly differ across continents. This suggests delays and cancellations are more likely due to airline operations or external factors rather than the departure continent.")

```
